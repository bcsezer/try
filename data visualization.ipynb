```python
countries_data = pd.read_csv("countries.csv")
```

```python
countries_data
```

```python
#örneğin amerika ve çinin popülasyonlarını karşılaştırıcaz
```

```python
tr = countries_data[countries_data.country == "Turkey"]
#burda olan şey ; countries_data ya eşitlediğimiz verilerin içinden,
#özel olarak bir veri serisini seçtik. countries_data adlı verininb country
#sütununda adı "Unided States" olanları getir dedik.
```

```python
fr = countries_data[countries_data.country == "France"]#united states için yaptğımız işlemin aynısı
```

```python
tr
```

```python
fr
```

```python
plt.plot(tr.year,tr.population / 10**6) #10**6 nın sebebi sayılar çok büyük,datanın daha iyi anlaşılabilmesi için m
plt.plot(fr.year,fr.population / 10**6)
plt.legend(["Turkey","France"])
plt.xlabel("YEAR")
plt.ylabel("POPULATİON")
plt.show()
```
